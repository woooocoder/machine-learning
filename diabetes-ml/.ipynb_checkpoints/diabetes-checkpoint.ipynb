# Extract

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

Diabetes = pd.read_csv('diabetes.csv') 
```

# Transform

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler 

cols = ['BMI', 'Insulin', 'BloodPressure', 'SkinThickness', 'Glucose']
for col in cols:
    Diabetes[col] = Diabetes[col].replace(0, np.nan)
 
other_cols = Diabetes.columns.difference(cols)   
scaled_data = Diabetes[cols]   

scaler = StandardScaler()  
scaled_data_scaled = scaler.fit_transform(scaled_data)

imputer = KNNImputer(n_neighbors=5, weights='uniform')
imputed_scaled_data = imputer.fit_transform(scaled_data_scaled)

scaled_data_imputed = scaler.inverse_transform(imputed_scaled_data)

DiabetesImputed = Diabetes.copy()
DiabetesImputed[cols] = pd.DataFrame(scaled_data_imputed, columns=cols).astype(Diabetes[cols].dtypes)

DiabetesImputed.head()
```

# Build train/test datasets

```python
from sklearn.model_selection import train_test_split 

# Develop train and test data and preprocess missing data. Pregnancies expects 0 values sometimes.
X = Diabetes[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']].values
y = Diabetes['Outcome'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=0, test_size=0.2)

X_imputed = DiabetesImputed[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']].values
y_imputed = DiabetesImputed['Outcome'].values
X_train_imputed, X_test_imputed, y_train_imputed, y_test_imputed = train_test_split(X_imputed, y_imputed, stratify=y_imputed, random_state=0, test_size=0.2)
```

# Cross Validation + Model Performance

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score, GridSearchCV 
parameters_ = {}

def cv(X_train, y_train, X_test, y_test):
    names = [
        'Logistic Regression',
        'KNN',
        'SVM',
        'Decision Tree',
        'Random Forest',
        'Neural Network',
    ]
    
    classifiers = [
        LogisticRegression(max_iter=1000),
        KNeighborsClassifier(),
        SVC(),
        DecisionTreeClassifier(),
        RandomForestClassifier(max_depth=4),
        MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42),
    ]

    res = {}

    for name, clf in zip(names, classifiers): 
        clf.fit(X_train, y_train)
 
        y_pred = clf.predict(X_test)
 
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)
        scores = cross_val_score(clf, X_train, y_train, cv=5)
        variance = scores.std() * 100 * 2
 
        res[name] = {
            'Accuracy': accuracy,
            'Variance': variance,
            'Precision': precision,
            'Recall': recall,
            'F1 Score': f1
        }

        parameters_[name] = {
            'params': clf.get_params()
        }
        
    for model_name, metrics in res.items():
        print("\nModel: %s" % model_name)
        for metric_name, metric_value in metrics.items():
            print("%15s: %0.2f" % (metric_name, metric_value)) 
cv(X_train_imputed, y_train_imputed, X_test_imputed, y_test_imputed)
```

## Tuning Hyperparameters + Model Performance

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
import numpy as np

# Dictionary to store hyperparameters and results
hyperparameters = {}

def tune_hyperparameters(X_train, y_train, X_test, y_test):
    names = [
        'Logistic Regression',
        'KNN',  
        'Random Forest', 
    ]
    
    classifiers = [
        LogisticRegression(max_iter=1000),
        KNeighborsClassifier(),
        RandomForestClassifier(max_depth=4), 
    ]
    
    param_grids = {
        'Logistic Regression': {
            'C': [0.1, 0.2, 0.5, 0.8, 1, 2, 3, 5, 7, 9, 10],
            'solver': ['liblinear', 'lbfgs'],
            'max_iter': [100, 500, 1000] 
        },
        'KNN': {
            'n_neighbors': np.arange(1, 50),
            'weights': ['uniform', 'distance'],
            'p': [1, 2]
        }, 
        'Random Forest': {
            'n_estimators': np.arange(1, 50, 5),
            'max_depth': [3, 5, 7],
            'min_samples_split': [2, 5],
            'min_samples_leaf': [1, 2],
            'max_features': ['sqrt'],
            'bootstrap': [False]
        }, 
    }
    
    best_params = {}
    for name, clf in zip(names, classifiers): 
        grid_search = GridSearchCV(clf, param_grids[name], cv=5, scoring='accuracy', n_jobs=-1)
        grid_search.fit(X_train, y_train)
        
        best_params[name] = {
            'best_params': grid_search.best_params_,
            'best_score': grid_search.best_score_,
            'score_std': grid_search.cv_results_['std_test_score'][grid_search.best_index_]
        }

        # Evaluate on test data
        best_model = grid_search.best_estimator_
        y_pred = best_model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred)
        recall = recall_score(y_test, y_pred)
        f1 = f1_score(y_test, y_pred)

        print(f"\nModel: {name}")
        print(f"Best Params: {grid_search.best_params_}")
        print(f"       Accuracy: {accuracy:.2f}")
        print(f"       Variance: {grid_search.cv_results_['std_test_score'][grid_search.best_index_] * 100:.2f}")
        print(f"      Precision: {precision:.2f}")
        print(f"         Recall: {recall:.2f}")
        print(f"       F1 Score: {f1:.2f}")
     
        hyperparameters[name] = {
            'params': grid_search.best_params_,
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'variance': grid_search.cv_results_['std_test_score'][grid_search.best_index_] * 100
        }

# tune_hyperparameters(X_train_imputed, y_train_imputed, X_test_imputed, y_test_imputed)
```

# Model Accuracy : Confusion Matrices

```python
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def evaluate_model(model, params):
    model.set_params(**params)

    model.fit(X_train_imputed, y_train_imputed)

    y_pred = model.predict(X_test_imputed)

    if hasattr(model, 'coef_'):
        coefficients = model.coef_[0]
        sorted_idx = coefficients.argsort()

#         plt.figure(figsize=(10, 7))
#         plt.barh(range(len(coefficients)), coefficients[sorted_idx], color='skyblue')
#         plt.yticks(range(len(coefficients)), ['BMI', 'Insulin', 'SkinThickness', 'Glucose', 'DiabetesPedigreeFunction', 'Age', 'Pregnancies', 'Blood Pressure'])
#         plt.xlabel('Coefficient Value')
#         plt.title(f'{model.__class__.__name__} Feature Importance')
#         plt.show()

    accuracy = accuracy_score(y_test_imputed, y_pred)
    print(f"Accuracy: {accuracy * 100:.2f}%")


    conf_matrix = confusion_matrix(y_test_imputed, y_pred)

    plt.figure(figsize=(6, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Diabetes', 'Diabetes'], yticklabels=['No Diabetes', 'Diabetes'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(str(model))
    plt.show()
    

# evaluate_model(KNeighborsClassifier(), {'n_neighbors': 47, 'weights': 'distance'})
# evaluate_model(KNeighborsClassifier(), hyperparameters.get('KNN').get('params'))
# evaluate_model(LogisticRegression(max_iter=5000), hyperparameters.get('Logistic Regression').get('params'))
# evaluate_model(RandomForestClassifier(), hyperparameters.get('Random Forest').get('params'))

evaluate_model(KNeighborsClassifier(), parameters_.get('KNN').get('params'))
evaluate_model(LogisticRegression(max_iter=5000), parameters_.get('Logistic Regression').get('params'))
evaluate_model(RandomForestClassifier(), parameters_.get('Random Forest').get('params'))

```
